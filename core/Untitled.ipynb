```python
def remove_suffix(text, suffix):
    return text[:-len(suffix)] if text.endswith(suffix) and len(suffix) != 0 else text

def remove_duplicates(x):
    a = []
    for i in x:
        if i not in a:
            a.append(i)
    return a
```

```python
import re as re

def extractCodeName(text):
    result = re.search('(\w*\s\d*\.)|(\w*\-\d*\.)|(\w*\s\d*\s)', text)
    result = result.group()
    
    result = remove_suffix(result, '.')
    result = remove_suffix(result, ' ')
    
    result = result.replace(' ', '-')
    
    print(result)
    
    return result

# extractCodeName('ОПК 7. Способен эксплуатировать технологич')
# extractCodeName('УК-9. Способен выполнять поиск, обработку')
```

```python
result = {}
curDis = frame[0][0]
curDisCodes = []

for i in range(0, frame.shape[0]):
    
    if frame[0][i] == curDis:
        curDisCodes.append(extractCodeName(frame[1][i]))
    else:
        result[curDis] = curDisCodes
        curDisCodes = []
    curDis = frame[0][i]
    
result[curDis] = curDisCodes

result
```
